### Comparando os resultados de cada método para cada uma das entradas do problema

#### *Entrada 1*

```python
# Médias dos dados fornecidos
metodos = ['Busca Exaustiva', 'Shuffle Fill', 'Random Selection']

# Médias dos tempos de execução, valores e pesos (substitua pelos seus dados reais)
media_tempo = [t_busca_exaustiva, avg_t_shuffle_1, avg_t_random_1]  # Tempo médio em segundos
media_valor = [v_busca_exaustiva, avg_v_shuffle_1, avg_v_random_1]  # Valor médio obtido
media_peso  = [p_busca_exaustiva, avg_p_shuffle_1, avg_p_random_1]  # Peso médio ocupado em kg

x = np.arange(len(metodos))  # Posições das barras

# Definir largura das barras
largura = 0.25

# Criar gráfico de barras
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# Gráfico de Tempo de Execução
ax[0].bar(x, media_tempo, width=largura, color='skyblue')
ax[0].set_title('Média do Tempo de Execução')
ax[0].set_ylabel('Tempo de Execução (s)')
ax[0].set_xticks(x)
ax[0].set_xticklabels(metodos)

# Gráfico de Valor Obtido
ax[1].bar(x, media_valor, width=largura, color='lightgreen')
ax[1].set_title('Média do Valor Obtido')
ax[1].set_ylabel('Valor Obtido')
ax[1].set_xticks(x)
ax[1].set_xticklabels(metodos)

# Gráfico de Peso Ocupado
ax[2].bar(x, media_peso, width=largura, color='salmon')
ax[2].set_title('Média do Peso Ocupado')
ax[2].set_ylabel('Peso Ocupado (kg)')
ax[2].set_xticks(x)
ax[2].set_xticklabels(metodos)

plt.tight_layout()
plt.show()
```

Observando os gráficos de tempo de execução, a média do valor obtido e média do peso ocupado, podemos perceber que o método de força bruta é de longo o mais lento mas também o mais preciso por uma margem relevante. Os métodos de *shuffle fill* e *random select* são muito mais rápidos (os valores de tempo desses dois métodos são tão pequenos que nem aparecem no gráfico dada a escala que foi definida para o eixo dos gráficos), mas também consideravelmente menos precisos. Por fim, é possível observar que todos os métodos apresentam um comportamento semelhante em relação ao peso ocupado.
