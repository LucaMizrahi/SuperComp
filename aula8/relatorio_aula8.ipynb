#### *Conclusão*

Ao observar o desempenho dos métodos de Shuffle Fill, Random Selection, Substituição de Objeto e Hill Climbing para as 3 entradas testadas nas aulas 7 e 8. O que é possível perceber é que o método de Shuffle Fill e Random Selection são os mais rápidos dado que esses métodos tomam decisões baseadas em critérios simples como aleatoriedade e probabilidade e não exploram todas as combinações possíveis. Já o método de Substituição de Objeto se demonstrou o mais lento entre os métodos (mas ainda extremamente mais rápido que o método de busca exaustiva), dado que este método busca otimizar a solução inicial por meio de substituições repetidas. Por fim o método de Hill Climbing se mostrou mais rápido que o método de Substituição de Objeto, dado que este método busca otimizar a solução inicial por meio de movimentos locais, ou seja, ele não explora todas as combinações possíveis, mas sim as combinações que estão próximas da solução inicial, o que pode ser eficiente para problemas de tamanho moderado.

Já em relação aos valores obtidos, o método de substituição foi consistente em obter valores maiores para todas as entradas testadas e essa diferença foi relevante, especialmente na entrada 1, chegando a ganhos de quase 100% quando comparado aos valores obtidos pelo shuffle fill por exemplo. Sendo assim, é possível perceber que o método de substituição seria uma boa alternativa para obter bons valores de resultado sem a necessidade de um tempo de execução muito longo, como ocorre com o método de busca exaustiva, que realmente consegue chegar a valores mais altos que todos os outros métodos testados, mas a um custo de tempo muito alto.

Por fim, em relação aos pesos, a diferença entre os pesos médios obtidos pelos métodos em cada entrada foi bem menos significante que a diferença nos valores obtidos, apesar de haver uma certa diferença em situações pontuais, como o método de substituição na entrada 2 e o método de random selection na entrada 4. Tirando esses casos, os pesos médios obtidos pelos métodos foram bem próximos, o que indica que a diferença nos valores obtidos pelos métodos não se deve a uma diferença nos pesos, mas sim a uma diferença na seleção dos objetos que compõem a mochila.
